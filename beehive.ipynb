```python
from ColabTurtle.Turtle import *
initializeTurtle(initial_window_size=(1000,600)) 
bgcolor('white')
color("black")
speed(13)
width(5)
up()
setpos(500,500)
down()
def hexgon():
  for counter in range(1,7):
    forward(20)
    left(60)
def top(amount_hex):
  for counter in range(1,amount_hex):
    hexgon()
    face(0)
    up()
    forward(40)
    down()
    face(270)
top(11)
up()
setpos(480,467)
down()
top(12)
up()
setpos(460,440)
down()
top(13)
up()
setpos(440,414)
down()
top(14)
up()
setpos(460,380)
down()
top(13)
up()
setpos(480,354)
down()
top(12)
up()
setpos(500,320)
down()
top(11)
```
