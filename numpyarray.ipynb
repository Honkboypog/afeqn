```python
import numpy as np

a = np.ones((5,6))
b = np.ones((4,3))
c = np.ones((3,4))
d = np.ones((4,3))
e = np.ones((3,4))

length = input("What array? ")

if length == "a":
  print(len(a))
elif length == "b":
  print(len(b))
elif length == "b":
  print(len(c))
elif length == "b":
  print(len(d))
elif length == "b":
  print(len(e))
else:
  print("INVALID 3RORR")

dimension = input("What array to dimesionate ")

if dimension == "a":
  print(a.ndim)
elif dimension == "b":
  print(b.ndim)
elif dimension == "c":
  print(c.ndim)
elif dimension == "d":
  print(d.ndim)
elif dimension == "e":
  print(e.ndim)
else:
  print("INV4L1D 3RR0R")

noelem = input("What array to elementante ")

if noelem == "a":
  print(a.size)
elif noelem == "b":
  print(b.size)
elif noelem == "c":
  print(c.size)
elif noelem == "d":
  print(d.size)
elif noelem == "e":
  print(e.size)
else:
  print("INV4L1D 3RR0R")

typelem = input("What array type elementante ")

if typelem == "a":
  print(a.dtype)
elif typelem == "b":
  print(b.dtype)
elif typelem == "c":
  print(c.dtype)
elif typelem == "d":
  print(d.dtype)
elif typelem == "e":
  print(e.dtype)
else:
  print("INV4L1D 3RR0R")

nametype = input("What array to nametype ")

if nametype == "a":
  print(a.dtype.name )
elif nametype == "b":
  print(b.dtype.name )
elif nametype == "c":
  print(c.dtype.name )
elif nametype == "d":
  print(d.dtype.name )
elif nametype == "e":
  print(e.dtype.name )
else:
  print("INV4L1D 3RR0R")
  
sum_wise = input("What array to sum ")

if sum_wise == "a":
  print(a.sum())
elif sum_wise == "b":
  print(b.sum())
elif sum_wise == "c":
  print(c.sum())
elif sum_wise == "d":
  print(d.sum())
elif sum_wise == "e":
  print(e.sum())
else:
  print("INV4L1D 3RR0R")

min = input("What array to find min ")

if min == "a":
  print(a.min())
elif min == "b":
  print(b.min())
elif min == "c":
  print(c.min())
elif min == "d":
  print(d.min())
elif min == "e":
  print(e.min())
else:
  print("INV4L1D 3RR0R")

max = input("What array to find min ")

if max == "a":
  print(a.max(axis=0))
elif max == "b":
  print(b.max(axis=0))
elif max == "c":
  print(c.max(axis=0))
elif max == "d":
  print(d.max(axis=0))
elif max == "e":
  print(e.max(axis=0))
else:
  print("INV4L1D 3RR0R")

mean = input("What array to find mean ")

if mean == "a":
  print(a.mean())
elif mean == "b":
  print(b.mean())
elif mean == "c":
  print(c.mean())
elif mean == "d":
  print(d.mean())
elif mean == "e":
  print(e.mean())
else:
  print("INV4L1D 3RR0R")
```
