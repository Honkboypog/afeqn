```python
from ColabTurtle.Turtle import *
initializeTurtle(initial_window_size=(1000,600)) 
bgcolor('white')
color("black")
speed(13)
width(5)
def body_line():
  forward(150)

def wheel():
  width(10)
  for counter in range(1,180):
    forward(2)
    left(2)
  
  left(90)
  for counter in range(4):
    width(5)
    forward(115)
    left(90)
    for counter in range(1,45):
      forward(1)
      left(1)
    left(90)
up()
setpos(700,300)
down()
wheel()
left(90)
for counter in range(1,68):
  forward(1)
  right(1)
left(100)
width(10)
body_line()
face(0)
forward(50)
for counter in range(1,90):
  forward(2)
  right(10)
up()
forward(40)
face(180)
down()
body_line()
face(0)
body_line()
right(120)
body_line()
right(120)
body_line()
forward(50)
face(180)
forward(50)
left(180)
forward(50)
right(60)
forward(50)
right(70)
body_line()
face(270)
wheel()
width(5)
forward(100)
body_line()
```
